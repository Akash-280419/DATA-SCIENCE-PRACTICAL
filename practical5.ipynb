```python
import pandas as pd
#Creating a datframe
df=pd.DataFrame({'Index':['S1','S2','S3','S4','S5','S6'],
                 'School_code':['S001','S002','S003','S001','S002','S004'],
                 'Name':['Akash ','Ritik','Abhinay','Sarthak','Rohan','Sagar'],
                 'Class':['V','V','VI','VI','V','V1'],
                 'Age':[18,18,19,19,18,19],
                 'Height':[173,182,193,178,190,185],
                 'Weight':[48,50,51,47,46,49],
                 'Address':['Nerul','Panvel','Vashi','Panvel','Kharghar','Panvel']})
df.set_index('Index',inplace=True)
df
```

```python
multiple_groups=df.groupby(['School_code','Class']).agg({'Height':['mean','min','max'],'Weight':['sum','min','count']})
print(multiple_groups)

```
